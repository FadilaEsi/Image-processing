```python
import os
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('resnet50.h5')

# Define the classes
classes = ['Normal', 'COVID19', 'pneumonia']

# Define the image size and channel
img_width, img_height, channel = 224, 224, 3

# Define the path to the test folder
test_dir = 'C:\\Users\\PC\\OneDrive\\Bureau\\image processing\\Data\\test'

# Loop through the classes
for class_name in classes:
    class_path = os.path.join(test_dir, class_name)

    # Loop through the images
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)

        # Load the image and resize it
        img = cv2.imread(img_path)
        img = cv2.resize(img, (img_width, img_height))

        # Preprocess the image
        img = img.astype('float32')
        img /= 255.0
        img = np.expand_dims(img, axis=0)

        # Make a prediction
        prediction = model.predict(img)

        # Print the result
        print('Image:', img_name)
        print('Class:', class_name)
        print('Prediction:', classes[np.argmax(prediction)])
        print()
```
